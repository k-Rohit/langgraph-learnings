```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
```

```python
# creating a namespace in the memory store - 
namespace = ('users', 'u1')
```

```python
# Adding memories
store.put(namespace=namespace, key="1", value={"data": "User likes pizza"})
store.put(namespace, "2", {"data": "User prefers dark mode"})
```

```python
# create another namespace
namespace2 = ('users', 'u2')
```

```python
# Adding memories to the newly created namespace
store.put(namespace=namespace2, key="1", value={"data": "User likes sushi"})
store.put(namespace2, "2", {"data": "User prefers grid style navigation"})
```
